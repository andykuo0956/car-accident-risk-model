```python
columns_lists = [
    list(df_2018A1.columns),
    list(df_2019A1.columns),
    list(df_2020A1.columns),
    list(df_2021A1.columns),
    list(df_2022A1.columns),
    list(df_2023A1.columns)
]


# 創建一個 dataframe，並將列表轉置
df = pd.DataFrame(columns_lists).T

# 為 dataframe 設置列名
df.columns = ["2018", "2019", "2020", "2021", "2022", "2023"]

# 輸出 dataframe
df
```

```python
df_A1_rawData = pd.concat([df_2018A1, df_2019A1, df_2020A1, df_2021A1, df_2022A1, df_2023A1])
df_A1_rawData = df_A1_rawData.dropna(subset=["發生日期"])
```

```python
df_A1_rawData
```

```python
df_A1 = df_A1_rawData
```

```python
df_A1[['死亡','受傷']] = df_A1['死亡受傷人數'].str.split(';', expand=True)
df_A1['死亡'] = df_A1['死亡'].apply(lambda x: re.sub('[^\d]+', '', x))
df_A1['受傷'] = df_A1['受傷'].apply(lambda x: re.sub('[^\d]+', '', x))
df_A1.drop(columns=['死亡受傷人數'], inplace=True)
```

```python
df_A1["發生月份"] = df_A1["發生月份"].astype(int)
df_A1["發生日期"] = pd.to_datetime(df_A1["發生日期"], format='%Y%m%d').dt.date
```

```python
df_A1
```

```python
df_A1_output = df_A1[["發生日期", "當事者區分-類別-大類別名稱-車種", "當事者區分-類別-子類別名稱-車種", "當事者屬-性-別名稱", "死亡", "受傷", "經度", "緯度"]]
```

```python
df_A1_output.columns = ['date', 'car_1', 'car_2', 'gender', 'death', 'injury', 'lng', 'lat']
```

```python
df_A1_output
```

```python
df_A1_output.to_csv("/Users/guoyanting/Desktop/NTUDAC/專案/Gogoro/rawData/traffic_accident/traffic_accident_A1.csv",index=False)
```

```python
df_A1_output["date"] = df_A1_output["date"].astype(str)
df_A1_output["date"] = df_A1_output["date"].str.replace("-", "")
```

```python
# 建立 GeoDataFrame
gdf = gpd.GeoDataFrame(
    df_A1_output, geometry=gpd.points_from_xy(df_A1_output['lng'], df_A1_output['lat']))

# 轉換座標系統，這裡以 EPSG 4326 為例
gdf = gdf.set_crs(epsg=4326)
```
